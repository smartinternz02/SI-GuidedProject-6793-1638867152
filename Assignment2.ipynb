```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = sns.load_dataset("titanic")
```

```python
data
```

```python
data.isnull().sum()
```

```python
data.drop("deck",axis=1,inplace=True)
```

```python
data.head()
```

```python
data["embarked"].isnull().any()
```

```python
data["embarked"].mode()
```

```python
missing_values=data[data.isnull().any(axis=1)]
```

```python
s=missing_values[missing_values["embarked"].isnull()]
```

```python
s
```

```python
s.index.tolist()
```

```python
data.drop(s.index.tolist(),axis=0,inplace=True)
```

```python
data["embarked"].isnull().sum()
```

```python
data.reset_index()
```

```python
data["age"].describe()
```

```python
gp_a=data.groupby("who")
```

```python
a=missing_values[missing_values["age"].isnull()]
```

```python
a.head()
```

```python
data["who"].unique()
```

```python
for id,row in a.iterrows():
   x=row["who"]
   avg_age=gp_a["age"].mean()[x]
   data.loc[id,"age"]=int(avg_age) 
```
